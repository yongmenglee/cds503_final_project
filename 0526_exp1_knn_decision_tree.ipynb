# Experiment 1 (Lee Kar Choon)<br>
The choices of available machine learning algorithms are greatly dependent on the nature of a business problem. In this case, the original values of the attribute "user_rating" consist of a list of discrete values (ranging from 0 to 5), it will be difficult to apply a regression algorithm for the prediction task, since it may yield a predicted outcome beyond the range of rating value. This problem can be framed as a multiclass classification problem and data preprocessing will be conducted to group several discrete values into the same class to simplify the prediction task. A few choices of machine learning algorithms specialized for classification task as below will be studied: K-Nearest Neighbors (KNN), Decision Tree and Support Vector Machine (SVM). Then, the recursive process of training and fine-tuning parameters of the machine learning model built from each of these 3 algorithms will be conducted. Finally, the machine learning algorithm that yields the best performing classification model will be identified and further investigated.

```python
# Indicate the target column
target = df['user_rating_label']
features = df.drop('user_rating_label', axis = 1)
#features = df.drop(['user_rating_label','ver'], axis = 1)
#target.head(10) #more grouping result
#features.head(10) #more grouping result
#target.tail(30) #more random result
#features.tail(30) #more random result
```

```python
target.value_counts()
```

```python
# Import train_test_split function 
# Import function for k-fold cross validation
# Import scikit-Learn metrics module for accuracy calculation
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

# Split data into train, validation and test sets
x_train, x_test, y_train, y_test = train_test_split(features, target, \
                                                   test_size = 0.2, random_state =0)

#K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier
# Create a KNN classifier with k = 5 
knn = KNeighborsClassifier(n_neighbors = 3)
scores = cross_val_score(knn, x_train, y_train, cv = 10, scoring = 'f1_macro')

knn.fit(x_train, y_train)
# Predict the target for the test dataset
test_predict = knn.predict(x_test)
# Display the array containing accuracy from 10 folds or iterations
# Print the mean accuracy score
# Compute the model accuracy on the test set
print(scores)
print('fl_macro (Validation) = ', scores.mean())
print('fl_macro (test): ',  metrics.f1_score(y_test, test_predict, average = 'macro'))
#print('Accuracy (test): ',  metrics.accuracy_score(y_test, test_predict))
# Print confusion matrix 
print(confusion_matrix(y_test, test_predict))
# Print precision, recall and F1 
print(classification_report(y_test, test_predict))

#param_norm_knn = {'n_neighbors': range(1,10),
#             'weights': ['uniform','distance'],
#             'p': [1,2] }
```

```python
# Try KNN standardized
from sklearn import preprocessing

names = features.columns

# Create the Scaler object
scaler = preprocessing.StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_features = pd.DataFrame(scaled_features, columns = names)
scaled_features.head()
```

```python
# Split data into train, validation and test sets
scaled_x_train, scaled_x_test, scaled_y_train, scaled_y_test = train_test_split(scaled_features, target, \
                                                   test_size = 0.2, random_state =0)
knn = KNeighborsClassifier(n_neighbors = 3)
scores = cross_val_score(knn, scaled_x_train, scaled_y_train, cv = 10, scoring = 'f1_macro')

knn.fit(scaled_x_train, scaled_y_train)
# Predict the target for the test dataset
test_predict = knn.predict(scaled_x_test)
# Display the array containing accuracy from 10 folds or iterations
# Print the mean accuracy score
# Compute the model accuracy on the test set
print(scores)
print('fl_macro (Validation) = ', scores.mean())
print('fl_macro (test): ',  metrics.f1_score(scaled_y_test, test_predict, average = 'macro'))
#print('Accuracy (test): ',  metrics.accuracy_score(y_test, test_predict))
# Print confusion matrix 
print(confusion_matrix(scaled_y_test, test_predict))
# Print precision, recall and F1 
print(classification_report(scaled_y_test, test_predict))

```

```python
# Import GridSearch from the sklearn library
from sklearn.model_selection import GridSearchCV

# Thep param_grid specifies one grid should be explored
param_grid = {'n_neighbors' : range(2,20)}

# Fit on the dataset on all parameter combinations in param_grid
# Retain the best combination
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv = 10, scoring = 'f1_macro') #f1_macro
grid_result = grid_search.fit(x_train, y_train)
grid_result.best_params_
#f1_macro: 3
#f1_weighted: 10
#accuracy: 18
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion = 'gini',splitter = 'best', random_state = 0)
scores = cross_val_score(dtree, x_train, y_train, cv = 10, scoring = 'f1_macro')

dtree.fit(x_train, y_train)
# Predict the target for the test dataset
test_predict = dtree.predict(x_test)
# Display the array containing accuracy from 10 folds or iterations
# Print the mean accuracy score
# Compute the model accuracy on the test set
print(scores)
print('fl_macro (Validation) = ', scores.mean())
print('fl_macro (test): ',  metrics.f1_score(y_test, test_predict, average = 'macro'))
#print('Accuracy (test): ',  metrics.accuracy_score(y_test, test_predict))
# Print confusion matrix 
print(confusion_matrix(y_test, test_predict))
# Print precision, recall and F1 
print(classification_report(y_test, test_predict))
```

```python
# Thep param_grid specifies one grid should be explored
param_grid = {'criterion' : ['gini','entropy'],
                    'splitter' : ['best','random']}
# Fit on the dataset on all parameter combinations in param_grid
# Retain the best combination
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv = 10, scoring = 'f1_macro') #f1_macro
grid_result = grid_search.fit(x_train, y_train)
grid_result.best_params_
```
